```python
# üîßÔ∏è Fix de Indenta√ß√£o (TabError)
print(f"Substitui cada Tab por 4 espa√ßos e remove espa√ßos in√∫teis no fim da linha")
print(f". . .")
def fix_indentation(filepath):
    if os.path.exists(filepath):
        with open(filepath, 'r') as f:
            lines = f.readlines()
        with open(filepath, 'w') as f:
            for line in lines:
                f.write(line.replace('\t', '    ').rstrip() + '\n')
        print(f"‚úÖ Arquivo corrigido: {filepath}")

fix_indentation('/content/vaultstream/core/pipeline.py')
fix_indentation('/content/vaultstream/ui/gradio_app.py')
fix_indentation('/content/vaultstream/notebooks/vaultstream_colab.ipynb')
fix_indentation('/content/vaultstream/core/encrypt.py')
fix_indentation('/content/vaultstream/core/uploader.py')
```
